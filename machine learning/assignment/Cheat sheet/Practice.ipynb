```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Create dummy data
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = (X[:,0] > 5).astype(int)

# Logistic regression model
model = LogisticRegression().fit(X, y)

# Plot decision boundary
X_plot = np.linspace(0, 10, 1000).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:,1]

plt.scatter(X, y, c=y, cmap='bwr')
plt.plot(X_plot, y_prob, color='green')
plt.axhline(0.5, color='black', linestyle='--')
plt.xlabel('Feature X')
plt.ylabel('Probability')
plt.title('Logistic Regression Decision Boundary')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Dummy Data
np.random.seed(42)
X = np.vstack([
    np.random.normal([0,0], 0.5, (50,2)),
    np.random.normal([5,5], 0.5, (50,2))
])

# KMeans clustering
kmeans = KMeans(n_clusters=2).fit(X)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='bwr')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='yellow', marker='X')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('K-Means Clustering')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy data simulation
np.random.seed(0)
X = np.linspace(0, 10, 100)
y_true = np.sin(X)
y_noisy = y_true + np.random.normal(0, 0.5, X.shape)

# Fit polynomial models of different complexity
degrees = [1, 3, 9]
plt.scatter(X, y_noisy, label='Data', color='gray', alpha=0.6)
plt.plot(X, y_true, label='True Function', color='black')

for deg in degrees:
    coeffs = np.polyfit(X, y_noisy, deg)
    y_pred = np.polyval(coeffs, X)
    plt.plot(X, y_pred, label=f'Degree {deg}')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```
