```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import seaborn as sns
```

```python
gdp_df = pd.read_csv('/Users/aran/Desktop/Semester_2/Introuction to Data Vizualisation/assignment 3/gdp_clean.csv')
print(gdp_df.head())
print(gdp_df.describe())
```

```python
life_df = pd.read_csv('/Users/aran/Desktop/Semester_2/Introuction to Data Vizualisation/assignment 3/life_exp_clean.csv')
print(life_df.describe())
life_df.head()
```

```python
pop_df = pd.read_csv('/Users/aran/Desktop/Semester_2/Introuction to Data Vizualisation/assignment 3/pop_clean.csv')
print(pop_df.head())
print(pop_df.describe())
```

```python
country_df = pd.read_csv('/Users/aran/Desktop/Semester_2/Introuction to Data Vizualisation/assignment 3/countrycodes.csv')
print(country_df.head())
print(country_df.describe())
```

```python
years = [1850,1900,1950,2000]
```

```python
merge = gdp_df.melt(id_vars = ["Country"], var_name="year", value_name = "GDP_per_capita")
merge = merge.merge(life_df.melt(id_vars=["Country"], var_name = "year", value_name ="Life_expectancy"),  on=["Country","year"])
merge = merge.merge(pop_df.melt(id_vars=["Country"], var_name="year", value_name="Population"),
                            on=["Country", "year"])
merge["year"] = merge["year"].astype(int)
```

```python
merge = merge.merge(country_df[["country", "continent"]],
                            left_on="Country", right_on="country", how="left")
merge.drop(columns=["country"], inplace=True)
```

```python
filtered= merge[merge["year"].isin(years)].copy()
continent_color = {
    "Asia": "red",
    "Europe": "green",
    "Africa": "black",
    "Oceania": "purple",
    "Americas": "blue"
}
filtered.head()
```

```python
filtered.loc[:, "Color"] = filtered["continent"].map(continent_color)

filtered.loc[:, "Bubble_size"] = filtered["Population"] / 100000 
filtered.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)
fig.suptitle("Life Expectancy vs Income: 1850-2000", fontsize=20 )

axes = axes.flatten()

for i, year in enumerate(years):
    ax = axes[i]
    data = filtered[filtered["year"] == year]

    scatter = ax.scatter(data["GDP_per_capita"], data["Life_expectancy"],
                         s=data["Bubble_size"],
                         c=data["Color"], alpha=0.6, edgecolors="black")

    ax.set_xscale("log")
    ax.set_title(str(year), fontsize=12)

    if i in [2, 3]:
        ax.set_xlabel("GDP/Capita $", fontsize=10)
    if i in [0, 2]:
        ax.set_ylabel("Life Expectancy in years", fontsize=10)

handle = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) 
           for color in continent_color.values()]
labels = continent_color.keys()
axes[3].legend(handle, labels, title="Continent", loc="lower right")
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

### Conclusion
The visualization successfully demonstrates how life expectancy and income evolved from 1850 to 2000.

A clear positive correlation is seen between GDP per capita and life expectancy—higher income generally leads to longer life expectancy.

The data disparity across continents is evident. In earlier years, most countries clustered at lower GDP and life expectancy, whereas in 2000, a wider distribution is seen with many nations improving significantly.
