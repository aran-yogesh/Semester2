```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")

flights_pivot = flights.pivot(index="month", columns="year", values="passengers") 

plt.figure(figsize=(10, 6))

sns.heatmap(
    flights_pivot,
    cmap="YlGnBu", 
    annot=True,  
    fmt="d", 
    linewidths=0.5, 
)

plt.title("Passenger Traffic Heatmap (Flights Dataset)")
plt.show()
```

This Seaborn heatmap uses cmap="YlGnBu" for a clear color gradient, annot=True to display exact values, and linewidths=0.5 for better separation. It’s a simple yet powerful way to visualize trends in tabular data
