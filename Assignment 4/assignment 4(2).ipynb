```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")


sns.set_style("darkgrid") 
sns.set_context("poster")  
sns.set_palette("muted")  

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
)


plt.title("Scatter Plot with Custom Aesthetics")

plt.show()
```

This sns scatterplot  visualizes the relationship between total bill and tip amount, with a customized aesthetic for improved readability.
