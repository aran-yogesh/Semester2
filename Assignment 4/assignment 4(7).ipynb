```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(
    data=iris, 
    hue="species",
    markers=["o", "s", "D"],  
    palette="husl", 
    height=2.5, 
)
plt.show()
```

This sns.pairplot showcases pairwise relationships using scatter plots and KDEs on the diagonal for distribution insights. The hue parameter differentiates categories with distinct colors and markers, making patterns and clusters easy to spot. A clean and effective way to explore feature relationships!
