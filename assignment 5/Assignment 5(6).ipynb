```python
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```python
df = px.data.iris()

fig = px.scatter_3d(
    df,
    x="petal_width",
    y="petal_length",
    z="sepal_length",
    color="species",
    title="3D Scatter Plot of Iris Dataset"
)
fig.show()
```

Problem 6.1.2

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
```

```python
fig = go.Figure(data=[
    go.Surface(
        z=z,
        x=x,
        y=y,
        colorscale="Viridis"
    )
])
fig.update_layout(
    title="3D Surface Plot of sin(sqrt(x² + y²))",
    scene=dict(
        zaxis_title="Z Value",
        xaxis_title="X Value",
        yaxis_title="Y Value"
    )
)
fig.show()
```

```python
gapminder_df = px.data.gapminder()
```

```python
fig = px.scatter_geo(
    gapminder_df[gapminder_df["year"] == 2007],
    locations="country",
    locationmode="country names",
    size="lifeExp",
    hover_name="country",
    title="Life Expectancy by Country (2007)",
    color="gdpPercap",
    projection="natural earth"
)
fig.show()
```
